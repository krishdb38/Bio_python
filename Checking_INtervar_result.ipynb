##### Function To check File Exists or Not 
Since We have different format file type and also we have many files so it is difficult to check so <br>
we pass case or control with required file format

````python
def check_file(case_or_control, file_extension = ".hg19_multianno.txt.intervar"):
    # "Annovar_result/1119/1119.hg19_multianno.txt.intervar"
    """This Function us useful to check weather file exists or Not and return 0 if 
    any file not exists"""
    file_extension = file_extension
    for _ in case_or_control:
        
        condition = os.path.exists('./Annovar_result/'+_+"/"+_+file_extension)
        if condition:
            print(_, "exist", end= "-->")
        else:
            print()
            print("\n")
            print("X-#-!"*20)
            print(_,"not exist")
            print("###"*20,"\n")
            print("Check this file and run",_)
            return 0
    print("\n\nAll File exists")
    return 1
````
