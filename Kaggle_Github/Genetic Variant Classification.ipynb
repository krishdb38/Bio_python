parse and encode the MC (Molecular Consequence) fild

```python
MC_list = df.MC.dropna().str.split(",").apply(lambda row: list(c.split("|")[1] for c in row))
MC_encoded = pd.get_dummies(MC_list.apply(pd.Series).stack()).sum(level = 0)
MC_encoded = MC_encoded.reindex(index = MC_list.index)

# Incorporate the transformed MC feature into the existinf Data Frame
df = df.join(MC_encoded).drop(columns=["MC"])

# Transformed MC feature
MC_encoded.head()
```

Manually generate crosstab, there is probably a faster method via pandas

```python
mccounts = {0:{},
           1:{},
           "All": {}
           }
for col in MC_encoded.columns:
    for class_ in [0,1]:
        mccounts[class_][col] = df.loc[df["CLASS"] == class_][col].sum()
    mccounts["All"][col] = df[col].sum

mc_ct = pd.DataFrame.from_dict(mccounts)

mc_ct_all = mc_ct.sum(axis = 0)
mc_ct_all.name = "All"
mc_ct = mc_ct.append(mc_ct_all, ignore_index = False)

mc_ct.drop("All", axis = 1, inplace = True)

mc_ct.plot.bar(stacked = True, figsize = (12,4))
```

Correlation for categorical features by way of chi-square test

```python
from itertools import combinations
from scipy.stats import chi2_contingency
```

```python
# Select a few categorical Features
categoricals_index = pd.MultiIndex.from_tuples(combinations(["CHROM", "REF", "ALT", "IMPACT","Consequence",\
                                                             "SYMBOL", "CLASS"],2))
categorical_corr = pd.DataFrame(categoricals_index, columns= ["cols"])
```

```python
def chisq_of_df_cols(row):
    c1, c2 = row[0], row[1]
    
    groupsizes = df.groupby([c1,c2]).size()
    ctsum = groupsizes.unstack(c1)
    
    # fillna (0) is necessary to remove any NAs which will cause exceptions
    return chi2_contingency(ctsum.fillna(0))[1]
```

```python
categorical_corr["chi2_p"] = categorical_corr.cols.apply(chisq_of_df_cols)
```

```python
categorical_corr
```

```python
categorical_corr.index = categoricals_index
categorical_corr = categorical_corr.chi2_p.unstack()
```

One of the ways variants can be classified is by the amount (and type) of sequence change.<br>
A substitution of a nucleotide (letter) is considered a single nucleotide variant (SNV), these are sometimes referred to as <br> **Single Nucleotide Polymorphisms(SNP).** <br><br>

When one or more nucleotides are inserted or deleted the variant is considered an insertion or deletion. There fore, if the <br> if the length of `REF` or `ALT` is >1 then the variant can be considered an Insertion or Deletion (indel), other wise it can be considered a SNV.

```python
snvs = df.loc[(df.REF.str.len() == 1) & (df.ALT.str.len()== 1)]
indels = df.loc[df.REF.str.len()>1 | (df.ALT.str.len()>1)]
```

```python
print(len(snvs) + len(indels))
print(len(df))

# Generally The len of SNV  Plus Indels will be the same
len(df) == len(snvs) + len(indels)
```

```python
df.loc[(df.REF.str.len() >1)]
```

```python
# SNVs are more likely to be conflicting than Indels
snp_indel =  pd.concat([
    snvs.CLASS.value_counts(normalize = True).rename("snv_class"),
    indels.CLASS.value_counts(normalize = True).rename("indel_class")],
    axis=1).T
```
