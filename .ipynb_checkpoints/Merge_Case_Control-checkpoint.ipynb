# This program will write final analysed file and merge all 10 files

```python
import pandas as pd
import numpy as np
import os
```

```python
data_files=os.listdir("../snp_analysis")
data_files[-1],len(data_files)
```

```python
case =[]
control =[]
#before running loop check the files and folders
for i in data_files:
    if i[-5]=="e":
        case.append(i)
    elif i[-5]=="l":
        control.append(i)
```

```python
#We know that there are 20 Files 
# For loop for all data sets
```

```python
len(control) #We can see control size is 11 so there is a folder
control
```

# One File of All case SNP

```python
df1 = pd.read_csv("../snp_analysis/"+case[0])
for data in case:
    print("Reading .."+data, end = "-->")
    if data == case[0]:
        pass
    df = pd.read_csv("../snp_analysis/"+data)
    df1 = df1.merge(df,how = "outer")
    df1 = df1.sort_values(by="POS")
df1.drop_duplicates()
df1.to_csv("../snp_analysis/merged_files/case_snp.csv",index=False,encoding='utf-8',header=True)
print("All Task Done")
    
#Remove duplicates rows later
```

# One Files of all Control 

```python
df1 = pd.read_csv("../snp_analysis/"+control[0])
for data in control:
    if data[-3:] != "csv":
        print(data+"skipped")
        pass
    print("Reading .."+data, end = "-->")
    if data == control[0]:
        pass
    df = pd.read_csv("../snp_analysis/"+data)
    df1 = df1.merge(df,how = "outer")
    df1 = df1.sort_values(by="POS")
df1.to_csv("../snp_analysis/merged_files/control_snp.csv",index=False,encoding='utf-8',header=True)
print("All Task Done")
```

# One Files of Case_Control

```python
df_case = pd.read_csv("../snp_analysis/merged_files/case_snp.csv")
df_control = pd.read_csv("../snp_analysis/merged_files/control_snp.csv")
df_all = df_case.merge(df_control,how = "outer")
df_all.to_csv("../snp_analysis/merged_files/case_control.csv",index=False,encoding='utf-8',header=True)
```
