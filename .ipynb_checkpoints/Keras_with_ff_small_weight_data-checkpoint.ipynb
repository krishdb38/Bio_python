### Load the dataset

```python
df = df[(df["GC_count"]>110.65) & (df["GC_count"]<112)]
df = df[df["FF"]<35]
```

```python
sns.pairplot(df)
```

```python
train_dataset = df.sample(frac = 0.8 , random_state=0)
test_dataset = df.drop(train_dataset.index)
```

```python
train_dataset.shape
```

#### Split the data into train and test

```python
train_data = df.sample(frac = 0.8 , random_state = 0)
test_dataset = df.drop(train_data.index)
```

```python
print(train_data.shape)
print(test_dataset.shape)
```

```python
train_stats = train_dataset.describe()
train_stats.pop("FF")
train_stats = train_stats.describe().T
```

```python
train_stats
```

```python
train_stats["mean"]
# mean and std are used for scalling our data 
# this also can be done with scikit library
```

```python
train_stats["std"]
```

#### Split feature that is Y Column

```python
train_labels = train_dataset.pop("FF")
test_labels = test_dataset.pop("FF")
```

```python
train_labels.head()
```

```python
test_labels.head()
```

```python
# Function to convert data to normalization
def norm(x):
    return (x- train_stats["mean"])/train_stats["std"]

normed_train_data = norm(train_dataset)
normed_test_data = norm(test_dataset)
```

```python
normed_test_data.head()
```

#### Build teh Model

Lets build our model.<br>
Here we'll use a Sequential model with two density connected hidden layers, and  an output layer that returns a single, continous value.

```python
print(len(train_dataset.keys()))
```

```python
def build_model():
    model = keras.Sequential([\
                             layers.Dense(64,activation= "relu", input_shape = [len(train_dataset.keys())]),
                             layers.Dense(64, activation = "relu"),
                             layers.Dense(1)])
    optimizer = tf.keras.optimizers.RMSprop(0.001)
    
    model.compile(loss = "mse", optimizer= optimizer, metrics= ["mae" , "mse"])
    return model
```

```python
model = build_model()
```

```python
#model.summary()
# We are using 2 hidden layers
```

#### Train the model

```python
EPOCHS = 1000
history = model.fit(\
                   normed_train_data, train_labels,\
                   epochs=EPOCHS , validation_split = 0.2,verbose=0,\
                   callbacks = [tfdocs.modeling.EpochDots()])
```

```python
test_predictions = model.predict(normed_test_data)
test_predictions[:10]
```
