```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import numpy as np
%matplotlib inline
```

# PCA_TRGC

```python
df = pd.read_csv('../datas_not_to_upload/trgc10000.csv', nrows=1000)
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.drop(columns = 'Unnamed: 0')
```

```python
df.shape
```

```python
df.head(3)
```

```python
df = df.loc[:, (df != 0).any(axis=0)]
```

```python
df
```

```python
df2 = pd.read_csv('data/bayesian_real/rgcbininfo_10000.txt', sep = '\s+', nrows=1000)

# Filter out grades that were 0
#df = df[~df['G3'].isin([0, 1])]
df2 = df2['B1']
#df = df.rename(columns={'G3': 'Grade'})

print(df2)
```

```python
from sklearn.preprocessing import StandardScaler


x = StandardScaler().fit_transform(df)
x
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=800)
principalComponents = pca.fit_transform(x)
rng = range(1, 801)
principalDf = pd.DataFrame(data = principalComponents, columns = ['V_' +str(i) for i in rng])
print(principalDf)
```

```python
finalDf = pd.concat ([principalDf, df2], axis = 1)
finalDf
```

```python
finalDf.to_csv("C:/Users/GC/data/bayesian_real/pca_trgc_2/pca_test800.csv", header=True, index=False)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)

targets = df2
colors = ['r', 'g', 'b']
for B1, color in zip(targets,colors):
    indicesToKeep = finalDf['B1'] == B1
    ax.scatter(finalDf.loc[indicesToKeep, 'V1']
               , finalDf.loc[indicesToKeep, 'V2']
               , c = color
               , s = 100)
ax.legend(targets)
ax.grid()
```

```python
pca.explained_variance_ratio_
```
