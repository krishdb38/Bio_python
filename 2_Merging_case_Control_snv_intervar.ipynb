```python
def split(x):
    "This Function will return only Front 1 or 2 word"
    splitted = x.rsplit()
    if splitted[0] in ["Pathogenic", "Benign"]:
        return splitted[:1][0]
    else:
        return "_".join(splitted[:2])
```
