```python
df2 = pd.read_csv('../datas_not_to_upload/gcbininfo_10000.csv', nrows=1000)
df2.head()
```

```python
from sklearn.preprocessing import StandardScaler

x = StandardScaler().fit_transform(df)
x
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=700) # Will reduce the number of Column
principalComponents = pca.fit_transform(x)
rng = range(1, 701)
principalDf = pd.DataFrame(data = principalComponents, columns = ['V_' +str(i) for i in rng])
principalDf.head(3)
```

```python
finalDf = pd.concat ([principalDf, df2], axis = 1)
finalDf.head()
```

```python
finalDf.to_csv("../datas_not_to_upload/pca_test700.csv", header=True, index=False)
```

# pca_tsrl

```python
df = pd.read_csv('../datas_not_to_upload/tsrl10000.csv', nrows=1000)
df.head()
```

```python
df = df.drop(columns = 'Unnamed: 0') # Remove Index COlumn Unnamed
```

```python
df = df.loc[:, (df != 0).any(axis=0)]
```

```python
df.head()
```

```python
df2 = pd.read_csv('../datas_not_to_upload/srlbininfo_10000', nrows=1000, sep="\t", header=None)
df2.columns = ["ID", "B1"]
df2.head()
```

```python
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(df)
x
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=700)
principalComponents = pca.fit_transform(x)
rng = range(1, 701)
principalDf = pd.DataFrame(data = principalComponents, columns = ['V_' +str(i) for i in rng])
principalDf.head()
```

```python
df2["B1"]
```

```python
finalDf = pd.concat ([principalDf, df2["B1"]], axis = 1)
finalDf.head()
```

### Save RL result

```python
finalDf.to_csv("../datas_not_to_upload/pca_rl_test700.csv", header=True, index=False)
```
